# 1 Using the parl.ch API to get a list of councillors

- Make a call to the API of the Swiss Parliament

```python
url = "http://ws-old.parlament.ch/councillors/basicdetails"
```

```python
params = {
    'councilFilter': 'N',
    'format': 'json'
}
```

```python
headers = {
    #"Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    #"Accept-Encoding":"gzip, deflate, sdch",
    #"Accept-Language":"de-DE,de;q=0.8,en-US;q=0.6,en;q=0.4",
    #"Cache-Control":"max-age=0",
    #"Connection":"keep-alive",
    #"Cookie":"PdInternet_UserLanguage=1031; SessionId=h12hr1vmvw1zeprqemgypd45; BIGipServerpool_frontend_ext_prod=177328394.20480.0000; ResponsiveImageRenditions_Viewport=827; PD_WebServices_Culture=de; TS011b62d7=0105a873d6f17cbf136d3289f1243e42998fb89bb238e7f2e453749777641dfdc6b5798537e77eb98697e051918b613d6e497606c714a4dd87fa9f14ebb56d12c23caf8ab8fdf37fde2f2a7695445f9cba276ab34f7e937d525b9f634daee533e3447aa46bef76ec57e2f59f8ad13745ca9e9cc727",
    #"Host":"ws-old.parlament.ch",
    #"Upgrade-Insecure-Requests":"1",
    "User-Agent":"Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}
```

```python
response = requests.get(url, params=params, headers=headers)
```

- Extract the needed information from the json and put it into a df

```python
data = response.json()
```

```python
data
```

```python
councillors = []
```

```python
for row in data:
    councillor = {}
    councillor['councillor_id'] = str(row['id'])
    councillor['firstName'] = row['firstName']
    councillor['lastName'] = row['lastName']
    councillor['canton'] = row['canton']
    councillor['party'] = row['party']
    councillors.append(councillor)
```

```python
df = pd.DataFrame(councillors, columns=['councillor_id', 'firstName', 'lastName', 'canton', 'party'])
```

```python
df
```

# 2 Save into the database

- Connect to our new database "parl"

```python
conn = pg8000.connect(database="parl", user="webapp", password="webapp")
```

```python
cursor = conn.cursor()
```

- prepare a function to insert the data and call it on the df

```python
def insert_councillor(row):
    query = "INSERT INTO councillors (councillor_id, firstName, lastName, canton, party) "
    query += "VALUES ('"+row.councillor_id+"', '"+row.firstName+"', '"+row.lastName+"', '"+row.canton+"', '"+row.party+"');"
    cursor.execute(query)
```

```python
df.apply(insert_councillor, axis=1)
```

```python
conn.commit()
```
