```python
import requests
import pandas as pd
import pg8000
```

# 1 Using the parl-API to get a list of all affairs

First, some basic info about the url

```python
url = "http://ws-old.parlament.ch/affairs"
```

```python
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}
```

```python
params = {
    'pageNumber': '',
    'format': 'json'
}
```

We will use this container to store all the affair_ids

```python
affairs = []
```

We define a function that takes json()-data and extracts the wanted info into the container

```python
def scrape_affair_ids(data):
    for row in data:
        current = {}
        current['affair_id'] = row['id']
        affairs.append(current)
```

Now we are ready to make repeated requests to the affair-list API.
We start go from page number 718 to 799 and scrape everything using the defined function

```python
for page_nr in range(718,800):
    params['pageNumber'] = page_nr
    response = requests.get(url, params=params, headers=headers)
    data = response.json()
    scrape_affair_ids(data)
```

```python
affairs[0]
```

```python
affairs[-1]
```

Success! Now we save the list into a dataframe

```python
df = pd.DataFrame(affairs, columns=['affair_id'])
```

# 2 Store the affair_id information into the database

Connect to the db

```python
conn = pg8000.connect(database="parl", user="webapp", password="webapp")
```

```python
cursor = conn.cursor()
```

We prepare a function that will store each row into the sql-database

```python
def insert_affair_id(row):
    query = "INSERT INTO affairs (affair_id) VALUES ('"+str(row.affair_id)+"');"
    cursor.execute(query)
```

We call this function

```python
df.apply(insert_affair_id, axis=1)
```

Commit the changes

```python
conn.commit()
```
