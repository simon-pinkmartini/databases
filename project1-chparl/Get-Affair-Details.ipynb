# 2 Using the parl-API to get detailed info on all affairs

First, some basic info about the url

```python
base_url = "http://ws-old.parlament.ch/affairs/"
```

```python
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}
```

```python
params = {
    'format': 'json'
}
```

We will use this container to store all the new information

```python
affairs = []
```

Prepare a function that will scrape the content for each API-call

```python
def scrape_affair_details(data):
    current = {}
    #this just to give a feeling of what's happening
    print ("ID:", data['id'], "Type:", data['affairType']['id'])
    #We are only looking for type-14-affairs
    if data['affairType']['id'] == 14:
        current['affair_id'] = str(data['id'])
        current['affair_type'] = str(data['affairType']['id'])
        current['councillor_id'] = str(data['author']['councillor']['id'])
        current['council'] = data['deposit']['council']['abbreviation']
        current['legislative_period'] = str(data['deposit']['legislativePeriod'])
        current['language'] = data['language']
        current['affair_title'] = data['title']
        #the json isn't always structured in the same way
        if data['texts'][0]['type']['name'] == "Antwort BR / BÃ¼ro":
            current['answer'] = data['texts'][0]['value'].strip()
            current['affair_text'] = data['texts'][1]['value'].strip()
        elif data['texts'][0]['type']['name'] == "Eingereichter Text":
            current['answer'] = ""
            current['affair_text'] = data['texts'][0]['value'].strip()
        else:
            current['answer'] = ""
            current['affair_text'] = ""
        affairs.append(current)
```

Make a series of API-Calls and use the scrape function on each call

```python
for affair_id in affair_ids:
#for affair_id in ['20154085', '20155182', '20155183', '20155195']: #this for testing
    url = base_url + affair_id
    response = requests.get(url, params=params, headers=headers)
    data = response.json()
    scrape_affair_details(data)
```

```python
len(affairs)
```

Save the affair list of dictionaries into a dataframe

```python
df = pd.DataFrame(affairs, columns=['affair_id', 'affair_type', 'councillor_id', 'council', 'legislative_period', 'language', 'affair_title', 'affair_text', 'answer'])
```

```python
df
```

```python
#Not the optimal way to deal with special characters but can't come up with other one.
df['affair_title'] = df['affair_title'].str.replace("'", "").str.replace('"', '')
df['affair_text'] = df['affair_text'].str.replace("'", "").str.replace('"', '')
df['answer'] = df['answer'].str.replace("'", "").str.replace('"', '')
```

Just for safety: save the df into a csv...

```python
df.to_csv('affairs.txt', sep="\t")
```

# 3 Store the new information into the sql-database

```python
conn = pg8000.connect(database="parl", user="webapp", password="webapp")
```

```python
cursor = conn.cursor()
```

Prepare a function to update the db row-by-row

```python
def update_affair(row):
    query = "UPDATE affairs SET "
    query += "affair_type = '" + row['affair_type'] + "', "
    query += "councillor_id = '" + row['councillor_id'] + "', "
    query += "council = '" + row['council'] + "', "
    query += "legislative_period = '" + row['legislative_period'] + "', "
    query += "language = '" + row['language'] + "', "
    query += "affair_title = '" + row['affair_title'] + "', "
    query += "affair_text = '" + row['affair_text'] + "', "
    query += "answer = '" + row['answer'] + "' "
    query += "WHERE affair_id = '" + row['affair_id'] + "';"
    cursor.execute(query)
```

Now, call this function on each row of the dataframe

```python
df.apply(update_affair, axis=1)
```

Commit the changes

```python
conn.commit()
```
